```python
import pandas as pd
```

```python
df = pd.read_excel('E-690.xlsx', usecols = 'A:J', nrows = 11)
df.head()
```

```python
attribute = (
    df.iloc[:, 0::2]
    .melt(value_name = 'Alphabets')
    .loc[:, ['Alphabets']]
)
attribute.head()
```

```python
value = (
    df.iloc[:, 1::2]
    .melt(value_name = 'V')
    .iloc[:, [1]]
)
value.head()
```

```python
result = (
    pd.concat([attribute, value], axis = 1)
    .groupby('Alphabets').agg(Sum = ('V', 'sum'))
    .reset_index()
)

result.head(5)
```
