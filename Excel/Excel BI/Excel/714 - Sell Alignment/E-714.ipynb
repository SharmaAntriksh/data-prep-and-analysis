```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("E-714 Sell Alignment.xlsx", usecols = 'A:B', header= 1, nrows = 10)
df
```

```python
df['Status'] = np.where(df['Produce'].str.contains('- N'), "Don't Sell", "Sell" )
df['Produce'] = df['Produce'].str.replace(' - N', '')
df
```

```python
pivot = (
    df
    .pivot_table(
        index = 'Farmer', columns = 'Status', 
        values = 'Produce', aggfunc = lambda x: ', '.join(sorted(x))
    )
    .reset_index()
    .fillna('')
)
pivot.columns.name = None
pivot[["Farmer", "Sell", "Don't Sell"]]
```
