```python
import pandas as pd
```

```python
df = pd.read_excel(r"E-724 Reverse alphabets and numbers separately.xlsx", usecols = 'A')
df
```

```python
def transform(text):

    def reverse(x):
        char, i = zip(*x)
        return list(zip(reversed(char), i))

    
    letters = [(char, i) for i, char in enumerate(text) if char.isalpha()]
    numbers = [(char, i) for i, char in enumerate(text) if char.isdigit()]

    if letters == []:
        result = reverse(numbers)
    elif numbers == []:
        result = reverse(letters)
    else:
        result = reverse(numbers) + reverse(letters)

    sort = sorted(result, key=lambda x: x[1])
    return ''.join([s for s, i in sort])
```

```python
df['Answer'] = df['Strings'].apply(transform)
df
```
