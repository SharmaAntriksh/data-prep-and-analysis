```python
import pandas as pd
```

```python
count = 7
start = 65

result = []

for i in range(count):
    
    char = chr(start + i)
    
    if i != count - 1: 
        result.extend([char] + [''] * (i + 1))
    else:
        result.extend(char)
        
result = pd.DataFrame([count] + result, columns = ['Answer'] )
result
```
