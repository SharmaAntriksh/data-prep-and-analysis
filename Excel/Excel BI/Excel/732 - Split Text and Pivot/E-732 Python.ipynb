```python
import pandas as pd
```

```python
df = pd.read_excel("E-732 Split Text and Pivot.xlsx", usecols = 'A', nrows = 4, header = 1)
df['Data'] = df['Data'].str.split(', ')
df = df.explode('Data')
df.head()
```

```python
df['Alphabet'] = df['Data'].apply(lambda x: x.split('-')[0])
df['Value'] = df['Data'].apply(lambda x: x.split('-')[1]).astype(float)
df = df.drop(columns = 'Data')
df.head()
```

```python
group = df.groupby('Alphabet')

temp = []

for n, g in group:
    
    g['I'] = ['Value' + str(i) for i in range(1, g.shape[0] + 1)]
    
    pivot = g.pivot_table(
        values = 'Value', index = 'Alphabet', 
        columns = 'I', aggfunc = 'sum'
    ).reset_index()

    temp.append(pivot)

result = pd.concat(temp, ignore_index = True)
result.columns.name = None
result = result.fillna(0)
result
```
