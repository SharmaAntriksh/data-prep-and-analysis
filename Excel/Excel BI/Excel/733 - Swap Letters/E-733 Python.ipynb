```python
import pandas as pd
```

```python
df = pd.read_excel('E-733 Swap Letters in Strings.xlsx', usecols = 'A:C', nrows = 10)
df['Rotation'] = df['Rotation'].astype(str)
df
```

```python
def rotate_letters(nums, words):

    s = words
    nums = map(int, nums.split(', '))
    
    for i in nums:
        if i >= 0:
            s = s[-i:] + s[:-i]
        else:
            s = s[i*-1:] + s[:i*-1]
    return s
```

```python
df['Rotation'] = df['Rotation'].astype(str)

df['Answer'] = df.apply(
    lambda row: rotate_letters(row['Rotation'], row['Words']), 
    axis = 1
)
df
```
