```python
import pandas as pd
```

```python
df = pd.read_excel("E-739 Count Vowels in All Substrings.xlsx", usecols = 'A:B', nrows = 9)
df
```

```python
def count_vowels(string: str):
    
    result = []
    vowels = 'aeiou'
    
    for i in range(len(string)+1):
        for k in range (i + 1, len(string) + 1):
            result.extend([string[i:k]])
    
    n = 0
    for word in result:
        for v in word:
            if v in vowels:
                n += 1
    return n
```

```python
df['Answer'] = df['String'].apply(count_vowels)
df
```
