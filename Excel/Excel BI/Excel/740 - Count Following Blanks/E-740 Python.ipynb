```python
df = pd.read_excel("E-740 Count Blanks.xlsx", usecols = 'A', skiprows = 1)
df['N'] = df['Data'].apply(lambda x: not pd.isna(x)).cumsum()

dfs = []

for n, group in df.groupby('N'):
    dfs.append(
        pd.DataFrame([[group.iloc[0,0], group.shape[0] -1]], columns = ['Alphabet', 'Count'])
    )

result = pd.concat(dfs, ignore_index = True)
result
```

```python
df = pd.read_excel("E-740 Count Blanks.xlsx", usecols = 'A', skiprows = 1)

df['N'] = df['Data'].notna().cumsum()
group = df.groupby('N')

result = pd.DataFrame(
    {
        'Alphabet': group['Data'].first(), 
        'Count': group.size() - 1
    }
)

result = result.reset_index(drop = True)
result
```
