```python
import pandas as pd
```

```python
data = pd.read_excel("E-750 Data Alignment.xlsx", usecols = 'A')
data.head(10)
```

```python
df = data.copy()
df['G'] = (df['Data'] == 'Emp ID').cumsum()
temp = []

for n, group in df.groupby('G'):

    vals = group['Data'].tolist()

    d = {
        'Emp ID': [v for v in vals if len(str(v)) == 4 and isinstance(v, int)] or [None],
        'Name': [v for v in vals if isinstance(v, str) and not v in ['Emp ID', 'Age', 'Name']] or [None],
        'Age': [v for v in vals if isinstance(v, int) and len(str(v)) <= 3] or [None]
    }
    temp.append(pd.DataFrame(d))

df = pd.concat(temp, ignore_index = True)
df
```

```python
df = data.copy()
df['G'] = (df['Data'] == 'Emp ID').cumsum()

temp = []

for n, group in df.groupby('G'):

    i = len(group) // 2
    d = dict(
        zip(
            group[:i]['Data'].tolist(), 
            group[i:]['Data'].tolist()
        )
    )
    temp.append(pd.DataFrame.from_dict([d]))

df = pd.concat(temp, ignore_index = True)
df.fillna('')
```
