```python
import pandas as pd
```

```python
data = pd.read_excel("E-801 Name and Seq.xlsx", usecols = 'A:B', header = 1, nrows = 5)
data
```

```python
df = data.copy()
df = (
    df
    .assign(x = df['Data'].str.replace('\t', '').str.split(';'))
    .explode('x')
    .pipe(
        lambda d: d.assign(**d['x']
                           .str.split(':', n = 1, expand = True)
                           .rename(columns = {0: 'Name', 1: 'Seq'})))
    .assign(Seq = lambda d: d['Seq'].str.split(','))
    .explode('Seq')
    .assign(
        Name = lambda d: d['Name'].str.strip(),
        Seq = lambda d: d['Seq'].str.strip().astype(int)
    )
    .loc[:, ['Name', 'Seq', 'State']]
    .sort_values(by = 'Seq')
    .reset_index(drop = True)
)
df
```
