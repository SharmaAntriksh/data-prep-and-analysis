```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('PQ-290.xlsx', sheet_name = 'Sheet1', usecols = 'A:B', nrows = 11)
df['N'] = (df['Column1'] == 'Company').astype(int).cumsum()
df
```

```python
result = []
for n, group in df.groupby('N'):
    values = group['Column2']
    cols = group['Column1']

    temp_df = pd.DataFrame(
        {
            col: [val] 
            for col, val in zip(cols, values)
        }
    )

    result.append(temp_df)

df = pd.concat(result, ignore_index = True)
df
```

```python
df = df.fillna(0)
df['Profit'] = np.where(
    df['Profit'] == 0, 
    df['Revenue'] - df['Cost'] - df['Tax'], 
    df['Profit']
)

df.reset_index(drop = True)
df
```
