```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('PQ-291.xlsx', usecols = 'A:D')

df['N'] = np.where(df['Company'].isna(), 0, 1).cumsum()
df = df.fillna('')
df['SalesRT'] = df.groupby('N')['Sales'].cumsum()
df['Tar'] = df.groupby('N')['Target'].transform('first')

df.head(15)
```

```python
result = (
    df[df['SalesRT'] <= df['Tar']]
    .drop(['SalesRT', 'Tar', 'N'], axis = 1)
    .reset_index(drop = True)
)
result.head(10)
```
