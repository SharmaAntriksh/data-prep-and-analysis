```python
import pandas as pd
```

```python
df = pd.read_excel('217 - Table Transformation.xlsx', usecols = 'B:C', nrows = 15, header = 1)
df['GroupN'] = df['Column 1'].astype(str).str.contains('=').cumsum()
df.head()
```

```python
result = []

for n, group in df.groupby('GroupN'):

    current_group = group.reset_index(drop = True)
    product_name = current_group.loc[0, 'Column 1'].split('=')[1]

    data = current_group.iloc[2:].reset_index(drop = True)

    temp = pd.DataFrame(
        {
            'Product': product_name,
            'Date': pd.to_datetime(data['Column 1']).dt.date,
            'Quantity': data['Column 2'].values
        }
    )

    result.append(temp)

solution = pd.concat(result, ignore_index = True)
solution
```
