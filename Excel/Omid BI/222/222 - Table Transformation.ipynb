```python
import pandas as pd
```

```python
df = pd.read_excel("222 - Table Transformation.xlsx", usecols = 'A', header = 1, nrows = 15)
df.head()
```

```python
df['Cluster'] = pd.Series(range(len(df) + 1)) // 3

import pandas as pd

def detect_dtype(value):
    try:
        if float(value).is_integer():
            return 3
    except:
        pass

    try:
        pd.to_datetime(value, dayfirst=True)
        return 1
    except:
        pass

    return 2

df['Sort'] = df['Column 1'].apply(detect_dtype)
df.head()
```

```python
result = []
for name, group in df.groupby('Cluster'):
    group = group.sort_values(['Sort'])['Column 1']
    group = pd.DataFrame([group])
    group.columns = ['Date', 'Product', 'Quantity']
    result.append(group)

ans = pd.concat(result, ignore_index= True)
ans['Date'] = pd.to_datetime(ans['Date'], format = '%d/%m/%Y')
ans
```
