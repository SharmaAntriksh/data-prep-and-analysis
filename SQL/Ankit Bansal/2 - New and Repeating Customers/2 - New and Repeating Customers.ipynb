```python
import pandas as pd
```

```python
df = pd.read_excel('2 - New and Repeating Customers.xlsx', usecols = 'B:E', nrows = 11, header = 1, parse_dates= ['order_date'])
df.head()
```

```python
df['FirstPurchaseDate'] = df.groupby('customer_id')['order_date'].transform('min')
df['Flag'] = (df['order_date'] == df['FirstPurchaseDate']).astype(int)
df
```

```python
result = (
    df
    .groupby('order_date')
    .agg(
        NewCustomer = ('Flag', 'sum'), 
        RepeatCustomer = ('Flag', lambda x: sum(x.map({1:0, 0:1})))
    )
    .reset_index()
)
result.columns = ['Order Date', 'New Customer', 'Repeat Customer']
result
```
